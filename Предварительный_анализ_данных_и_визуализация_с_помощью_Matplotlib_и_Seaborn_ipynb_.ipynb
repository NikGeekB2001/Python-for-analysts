#### **Анализ красного португальского вина "Винью Верде**

```python
# Импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt #работает с графиками и диаграммами
import seaborn as sns #работает с графиками и диаграммами
```

```python
# Считываем датасет
df = pd.read_csv(('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'), sep = ';')
df.sample(n = 10)
```

```python
df.head() # Первые 5 строк

```

```python
df.tail()   # Последние 5 строк
```

### 2. На основе переменной quality с оценками вина создать новую переменную good: если quality > 5, то 1, иначе 0.

```python
 df['good'] = df['quality'] 
```

```python
df.head()
```

```python
# Для удоства последующих расчетов, перекодируем названия событий в 0/1 с помощью лямбда-функции
df['good'] = df['quality'].apply(lambda x: 1 if x > 5 else 0)
df.head(15)
```

### 3. Исследовать, какие химические характеристики вина влияют на то, окажется оно хорошим или плохим,  с применением не менее 5 диаграмм из урока.

```python
# Разведочный анализ - выявление основных структур/ Строим гистограмму в разрезе содержания спирта в португальском красном вине  "Винью Верде"
# Распределение переменной оценивают с помощью гистограмм. 
#Гистограмма - двухмерный график, по горизонтальной оси которого откладываются переменные или числовые интервалы, 
# а по вертикальной – частота появления переменной (в заданном интервале).

# Строим гистограмму в разрезе содержания спирта

plt.figure(figsize = (10,7))

plt.hist(df['alcohol'], bins = 120)  # поставил 120 бинов, для наглядности и визуализации выбросов

plt.title('Содержания спирта в португальском красном вине "Винью Верде')
plt.xlabel('Интервал, наличия спирта в кагортах')
plt.ylabel('Содержание спирта')

plt.show()

# Заключение: Из графика видно, что распределение переменной на интервал с числовым  интервалом (9,5 - 9,6). 
# Другими словами - содержание спирта (алкоголя) находится в пределах от 9.0 до 14.0 
```

```python
# Строим гистограмму в разрезе содержания остаточного сахара

plt.figure(figsize = (10,7))

plt.hist(df['residual sugar'], bins = 510)  # поставил 510 бинов, для наглядности и визуализации выбросов

plt.title('Содержания остаточного сахара в португальском красном вине "Винью Верде')
plt.xlabel('Интервал, наличия сахара в кагортах')
plt.ylabel('Содержание сахара')

plt.show()

# Заключение: Из графика видно, что распределение переменной на интервал с числовым  интервалом (2,0 - 2,1). 
# Другими словами - содержание остаточного сахара находится в пределах (2,0 - 2,1). 
# Актуально для диабетиков
```

```python
# Выводим данные для просмотра
df.head()
```

```python
# Используем функцию crosstab и ее параметр normalize для получения относительных частот 
pd.crosstab(index = df['total sulfur dioxide'], columns = df['good'], normalize='index')

# Заключение: общий разброс категории вина составляет от 0 до 1, разброс показателей общего диоксида серы составляет от 6.0 до 289.0
# После этого определяется частота появления всех возможных комбинаций обоих признаков в собранных данных и значение ее заносится в ячейки,
# стоящие на пересечении соответствующих строк и столбцов. 
# Совместное распределение категориальных переменных и позволяют выявить наличие (отсутствие) взаимосвязи между переменными.
# Показатели общего диоксида серы 1.000000 находятся в 4 категорях, данный факт означает, что большая часть диаксида серы находится в вине хорошего качества (с оценкой пользователей пять) 
```

```python
# Воспользуемся методом Groupby, для определения значений наличия остаточного сахара в вине по качеству (оценка пользователей от 0 до 10)
df_1 = df.groupby('residual sugar').agg({'quality':['min', 'max', 'mean']})
df_1
# Вывод: клиенты отдают больше предпочтение красному вину, где садержание сахара  от 0.9 до  15.4 (рассмотрено среднее значение)
# Выше 15.4 производить красное вино "бизнесу" экономически нецелесообразно..
```

```python
# Осуществляем вывод
df.info()
```

```python
# Смотрим ниже на сет и прикидываем интервалы total    
df_3 = df.groupby('total sulfur dioxide').agg({'good':['min', 'max', 'mean']})
df_3
```

```python
# разбивка на интервалы, как видим выше в категории 6_ общий диоксид серы, целое число
breaks = [0, 50.0, 100.0, 150.0, 250.0, 289.0]
df['total sulfur dioxide_bins'] = pd.cut(df['total sulfur dioxide'], bins = breaks).astype('str')
df['total sulfur dioxide_bins'].head()
```

```python
# Строим диаграмму рассеяния

plt.figure(figsize = (10,7))
plt.scatter(df['total sulfur dioxide_bins'], df['quality'])  
plt.title('График разброса значений общего диоксида серы')
plt.xlabel('Общий диоксид серы по категориям')
plt.ylabel('Качественная оценка пользователей от 0 до 10')

plt.show()

# Вывод: На графике видно, что на оценку пользователей приходится содержание общего диоксида серы от 0 до 100.0.
# Следовательно доводим до "бизнеса", что общий диоксид серы необходимо держать от 0 до 100.0.
# В дальнейшем категорию общего диоксида серы можно разложить подетальнее..
```

```python
# Строим столбчатую диаграмму

plt.figure(figsize = (10,7))

plt.bar(df['total sulfur dioxide_bins'],df['quality'])

plt.title('График разброса значений общего диоксида серы')
plt.xlabel('Общий диоксид серы по категориям')
plt.ylabel('Качественная оценка пользователей от 0 до 10')
plt.xticks(rotation = 30)

plt.show()

# Подтверждение данных диаграммы рассеяния
# Вывод: На графике видно, что на оценку пользователей приходится содержание общего диоксида серы от 0 до 100.0.
# Следовательно доводим до "бизнеса", что общий диоксид серы необходимо держать от 0 до 100.0.
# В дальнейшем категорию общего диоксида серы можно разложить подетальнее..
```

### **Анализ португальского красного вина**

```python
# При анализе данных необходимо обратить внимание на показатель - pH показатель кислотности, т.к. вино должно быть с легкой кислинкой
# Строим диаграмму рассеяния

plt.figure(figsize = (10,7))
plt.scatter(df['pH'], df['good'])  
plt.title('График разброса значений общего диоксида серы')
plt.xlabel('Общий диоксид серы по категориям')
plt.ylabel('Качественная оценка пользователей от 0 до 1')

plt.show()

# Вывод: активная кислотность сусла и вина колеблется в среднем в пределах рН 2,8 – 3,8.
# как мы видим на нашем графике разброс с оценкой пользователей более 5 составляет от 2.8 до 3.7,что является нормой.
# разброс с оценкой пользователей менее 5 составляет от 2.9 до 3.7,что является нормой.

# Подкислять вино лимонной кислотой  нет необходимости
```
